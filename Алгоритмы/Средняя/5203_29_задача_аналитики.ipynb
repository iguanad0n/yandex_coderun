В трек аналитики второго сезона вошло 28 задач, однако кроме них была подготовлена ещё одна задача. Так как в сезон она не вошла, то предлагаем решить вам её сейчас.

Вам дано квадратное уравнение вида:

ax^2 + bx + c = 0, где a,b и c — целые числа, такие что a≠0.

Напишите программу, которая решает это уравнение.

## **Формат ввода**
Единственная строка входных данных содержит три целых числа a,b,c(−10^3≤a,b,c≤10^3,a≠0).

## **Формат вывода**

В первой строке выведите количество действительных корней уравнения.

В второй строке выведите сами корни. Если уравнение имеет два корня, то выведите их через пробел в порядке возрастания. Если корень один, то выведите его. Если корней нет, то ничего выводить не нужно. Ваш ответ будет принят, если введенные числа отличаются от правильных ответов на величину не более 10 ** (-6).



```python
import sys


def main():
    a, b, c = map(int, input().split())
    dscr = b**2 - 4 * a * c
    if dscr > 0:
        x1 = (-b - dscr ** 0.5) / (2 * a)
        x2 = (-b + dscr ** 0.5) / (2 * a)
        print(2)
        print(*sorted([x1,x2]))
    elif dscr == 0:
        print(1)
        print((-b) / (2*a))
    else:
        print(0)


if __name__ == '__main__':
    main()
```
